Solusi / Saran yang dapat kami ajukan berdasarkan hasil pengelompokkan / clustering seperti gambar adalah : 
- Label 0 (Warna Merah) , merupakan sekumpulan customer dengan ciri memiliki annual income yang berada di kisaran 10-40k , dengan spending sccore 4 - 40
- Label 1 (Warna Biru), merupakan sekumpulan customer dengan ciri memiliki annual income yang berada di kisaran 70-135k, dengan spending score 60 - 100
- Label 2 (Warna Kuning), merupakan sekumpulan customer dengan ciri memiliki annual income yang berada di kisaran 70-135k, dengan spending score 0 - 40
- Label 3 (Warna Hijau), merupakan sekumpulan customer dengan ciri memiliki annual income yang berada di kisaran 40-70k, dengan spending score 36 - 60
- Label 4 (Warna Magenta) merupakan sekumpulan customer dengan ciri memiliki annual income yang berada di kisaran 10-40k , dengan spending score 60 - 100

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (14, 8))

plot_color = ['red', 'blue', 'orange', 'green', 'magenta']
marker_gen = {'Male':'^', 'Female':'s'}
for gen in customer['Gender'].unique():
    for i in customer['Label'].unique():
        label_gen_lab = customer[(customer['Gender'] == gen) & (customer['Label'] == i)]
        plt.scatter(label_gen_lab['Annual Income (k$)'], label_gen_lab['Spending Score (1-100)'] , color = plot_color[i] , marker=marker_gen[gen])
plt.title('Clustering Customer by Spending Score and Annual Income for Fast Moving Consumer Goods (FMCG)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')        
plt.legend(['Label 0','Label 4','Label 3','Label 1','Label 2'])
plt.show()
```

Dengan asumsi perusahaan yang dijalankan adalah FMCG (Fast Moving Consumer Goods) : 
- Label 0 , memiliki ciri dimana customer memiliki spending yang cukup rendah , yang didorong dengan penghasilan yang cukup rendah
- Label 1 , memiliki ciri dimana customer memiliki spending yang tinggi, juga didorong penghasilan yang tinggi , sehingga bisa dikatakan label 1 adalah customer yang cukup royal & konsumtif
- Label 2, memiliki ciri dimana customer memiliki spending yang rendah, namun memiliki annual income yang cukup tinggi, sehingga dapat dikatakan customer label 2 memiliki sifat 'pelit' / cukup 'pilih - pilih' dalam menggunakan uangnya
- Label 3 , memiliki ciri dimana customer memiliki spending & annual income berada pada rata - rata.
- Label 4, memiliki ciri dimana customer memiliki spending yang tinggi, namun tidak disertai annual income yang selaras . Dalam kata lain , customer ini cukup boros padahal penghasilannya rendah

Berdasarkan target yang cocok untuk dijalankan pada marketing dapat ditargetkan pada kelompok **Label 1** dengan berisikan customer yang memiliki penghasilan (penghasilan diatas 70k/tahun) dan pengeluaran banyak dengan disajikan produk yang berkualitas dan dengan harga yang relatif tinggi .Namun, dampaknya akan sulit untuk fokus pada kelompok ini karena income yang diperoleh akan lebih lama dibandingkan dengan jumlah orang pada label 3. 

Untuk **label 2**, dimana spendingnya yang cukup sedikit padahal penghasilan sangat tinggi , dapat kita beri stimulus berupa promo dengan paket hemat / value buy pada produk yang kita tawarkan. Sehingga, nantinya yang cukup pilih - pilih untuk membeli barang yang termurah , akan lebih tergiur oleh diskon / value buy dengan syarat tertentu yang membuat customer tersebut mau tidak mau akan spending (belanja barang lebih) agar mendapatkan diskon tersebut

Untuk label 1, customer ini cukup berharga jika kita lihat dari tingkat royalitasnya yang cukup tinggi (royalitas diukur berdasarkan tingkat konsumtif tinggi dan selaras dengan penghasilannya) , sehingga akan lebih baik jika kita memperhatikan customer pada label 1, dari sisi customer retention , sehingga customer ini tidak mudah pindah ke kompetitor. Salah satu solusi yang dapat diberikan kepada customer ini adalah sistem membership. Membership ini akan memperoleh privilege, berupa undian grand prize , diskon khusus , atau tempat parkir khusus. Namun , membership ini tentu tergantung bagaimana customer mampu melakukan spending yang tinggi / tidak.
