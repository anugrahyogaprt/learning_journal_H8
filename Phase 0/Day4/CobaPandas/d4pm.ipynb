```python
# year > 2100
# fran_id Warriors
# pts > 100
# game_result W

df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Warriors') &
    (df['pts'] >= 100) &
    (df['game_result'] == 'W')
][['year_id', 'fran_id', 'pts', 'opp_pts', 'opp_fran', 'game_result']]
```

```python
df[['year_id', 'fran_id', 'pts', 'opp_pts', 'opp_fran', 'game_result']][
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Warriors') &
    (df['pts'] >= 100) &
    (df['game_result'] == 'W')
]
```

```python
df[['year_id', 'fran_id', 'pts', 'opp_pts', 'opp_fran', 'game_result']][
    (df['year_id'] > 2010) &
    (df['year_id'] < 2013) &
    (df['fran_id'] == 'Warriors') &
    (df['pts'] >= 100) &
    (df['game_result'] == 'W')
]
```

```python
df.query("year_id > 2010 & fran_id == 'Warriors' & pts >= 100 & game_result == 'W'")
```
